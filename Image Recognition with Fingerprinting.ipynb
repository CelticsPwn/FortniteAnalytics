```python
import dhash
from PIL import Image
import glob

f = open("items" + " hashcode.txt", "w")

for filename in glob.glob('Items/*.png'):
    image = Image.open(filename)
    row, col = dhash.dhash_row_col(image)
    hash_num = dhash.format_hex(row, col)
    f.write(hash_num + " : " + filename + ",")
    print(hash_num + " : " + filename + ", \n")

f.close()
```

```python
import dhash
from PIL import Image

file = open("items" + " hashcode.txt", "r")
f = open("Weapon Names.txt", "r")
image = Image.open("Item-8.jpg")

row, col = dhash.dhash_row_col(image)
hash_str = dhash.format_hex(row, col)
hash = int(hash_str, 16)

lines = file.read().split(',')
differences = []
diction = {}

#get the names of the weapons
weapons = f.read().split(',')

#RGB values for rarity of weapons
legendary = (180, 105, 65)
epic = (130, 50, 175)
rare = (15, 150, 180)
uncommon = (85, 165, 10)
common = (140, 140, 140)

#compare all the hashcodes
for line in lines:
    if (len(line) > 32):
        hash_string = line[:32]
        hash_item = int(hash_string, 16)
        diff = dhash.get_num_bits_different(hash, hash_item)
        item_str = line[41:]
        item = item_str[:-8]
        differences.append(diff)
        diction[diff] = item 

#get the item with the closes hash code
num = min(differences)
item = diction.get(num)

#if item is weapon, get the rarity
if (item in weapons):
    pix = image.load()
    RGB = pix[10,10]
    rar = {}
    #get abs difference of RGB value
    legendary_val = abs(legendary[0]-RGB[0]) + abs(legendary[1]-RGB[1]) + abs(legendary[2]-RGB[2])
    rar[legendary_val] = "Legendary"
    epic_val = abs(epic[0]-RGB[0]) + abs(epic[1]-RGB[1]) + abs(epic[2]-RGB[2])
    rar[epic_val] = "Epic"
    rare_val = abs(rare[0]-RGB[0]) + abs(rare[1]-RGB[1]) + abs(rare[2]-RGB[2])
    rar[rare_val] = "Rare"
    uncommon_val = abs(uncommon[0]-RGB[0]) + abs(uncommon[1]-RGB[1]) + abs(uncommon[2]-RGB[2])
    rar[uncommon_val] = "Uncommon"
    common_val = abs(common[0]-RGB[0]) + abs(common[1]-RGB[1]) + abs(common[2]-RGB[2])
    rar[common_val] = "Common"
    
    min_val = min(legendary_val, epic_val, rare_val, uncommon_val, common_val)
    rarity = rar.get(min_val)
    item = item + " - " + rarity
    
print(item)
    
file.close()
f.close()
```
